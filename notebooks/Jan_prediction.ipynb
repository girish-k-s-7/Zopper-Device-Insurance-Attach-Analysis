### Zopper Device Insurance Attach Analysis :- January Prediction Modelling

```python
import pandas as pd
df = pd.read_excel("../data/raw/Jumbo_and_company_attach%.xlsx")

df.head()
```

```python
# Predicting January attach percentage using weighted average of last three months
df["Jan_Predicted_Attach"] = (
    0.5 * df["Dec"] +
    0.3 * df["Nov"] +
    0.2 * df["Oct"]
)
```

```python
# Saving relevant columns for January prediction
jan_prediction = df[["Branch", "Store_Name", "Jan_Predicted_Attach"]]

jan_prediction.head()
```

```python
# Saving outputs
jan_prediction.to_csv("../outputs/january_attach_prediction.csv", index=False)
```

January attach rates are projected to remain broadly consistent with recent performance, with higher-performing stores continuing to show stronger insurance adoption compared to lower-performing locations.
