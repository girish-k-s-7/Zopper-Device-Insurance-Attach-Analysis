### Zopper Device Insurance Attach Analysis :- Data Cleaning

```python
import pandas as pd
import os
df = pd.read_excel(r"C:\Users\giris\OneDrive\Documents\Jumbo_and_company_attach%.xlsx")
```

```python
df.head()
```

```python
df.info()
```

```python
# reshaping the data
df = df.melt(
    id_vars=["Branch", "Store_Name"],
    value_vars=["Aug", "Sep", "Oct", "Nov", "Dec"],
    var_name="Month",
    value_name="Attach_Percentage"
)

df.head()
```

```python
# month ordering
month_order = ["Aug", "Sep", "Oct", "Nov", "Dec"]
df["Month"] = pd.Categorical(
    df["Month"],
    categories=month_order,
    ordered=True
)
```

```python
df.shape
```

```python
# saving the data
output_path = "../data/processed/zopper_attach_Processed.csv"
df.to_csv(output_path, index=False)
```

**Conclusion**

The dataset was reshaped from a wide monthly format to a long, analysis-ready structure with one row per store per month. All attach percentage values were already numeric and required no missing value treatment. The resulting dataset ensures consistent month ordering and provides a standardized foundation for month-wise, branch-wise, and store-level performance analysis.
