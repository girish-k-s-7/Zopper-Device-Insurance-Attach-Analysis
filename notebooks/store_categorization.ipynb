### Zopper Device Insurance Attach Analysis :- Store Categorization

```python
import pandas as pd

df = pd.read_csv("../data/processed/zopper_attach_processed.csv")

store_stats = (
    df
    .groupby(["Branch", "Store_Name"])["Attach_Percentage"]
    .agg(
        Avg_Attach="mean",
        Std_Dev="std"
    )
    .reset_index()
)

store_stats.head()
```

**Categorizing store**

Rules:

High: Avg_Attach ≥ 0.40

Medium: 0.20 ≤ Avg_Attach < 0.40

Low: Avg_Attach < 0.20

Inactive: Avg_Attach = 0

```python
def categorize_store(avg):
    if avg == 0:
        return "Inactive"
    elif avg >= 0.40:
        return "High"
    elif avg >= 0.20:
        return "Medium"
    else:
        return "Low"

store_stats["Category"] = store_stats["Avg_Attach"].apply(categorize_store)

store_stats.head()
```

Stores were classified into high, medium, low, and inactive performance categories based on their average attach rates to enable targeted, store-specific improvement actions.
